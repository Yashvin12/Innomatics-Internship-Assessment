```python
# Problem Statement 1: Smart Parking Lot Management System

def manage_parking(parking_capacity, vehicle_logs):
    current_vehicles = 0      # Tracks vehicles currently parked
    peak_usage = 0            # Tracks maximum vehicles at any time
    # Loop through each log entry
    for entry in vehicle_logs:
        if entry == "IN":
            current_vehicles += 1
        elif entry == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1
        # Update peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)
    # Check capacity condition
    if current_vehicles > parking_capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

# Sample Run
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem Statement 2: Online Food Delivery Time Estimator
def estimate_delivery_time(distance_km, traffic_level, weather_condition):
    # Base time: assume 5 minutes per km
    estimated_time = distance_km * 5
    # Add delay based on traffic
    if traffic_level == "High":
        estimated_time += 15
    elif traffic_level == "Medium":
        estimated_time += 8
    
    # Add delay based on weather
    if weather_condition == "Rainy":
        estimated_time += 10
    elif weather_condition == "Storm":
        estimated_time += 20
    
    print("Estimated Delivery Time:", estimated_time, "minutes")
# Sample Run
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem Statement 3: Movie Theatre Seat Occupancy Analyzer
def analyze_seat_occupancy(total_seats, booked_seats_list):
    total_booked = len(booked_seats_list)
    occupancy_percent = (total_booked / total_seats) * 100
    
    print("Occupancy:", int(occupancy_percent), "%")
    if occupancy_percent == 100:
        print("Show Status: Housefull")
    elif occupancy_percent >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
analyze_seat_occupancy(200, [1] * 150)
```

```python
# Problem Statement 4: Cloud Server Load Classification
def classify_server_load(cpu_readings):
    total_cpu = 0
    # Calculate total using loop
    for value in cpu_readings:
        total_cpu += value
    average_cpu = total_cpu / len(cpu_readings)
    print("Average CPU Load:", int(average_cpu), "%")
    if average_cpu < 50:
        print("Server Status: Normal")
    elif average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
# Sample Run
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Problem Statement 5: Smart Classroom Resource Monitor

def monitor_classroom_resources(resource_usage):
    overused_items = []
    for resource, hours_used in resource_usage.items():
        if hours_used > 8:
            overused_items.append(resource)
    if len(overused_items) > 0:
        print("Overused Resources:", ", ".join(overused_items))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
monitor_classroom_resources({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Problem Statement 6: Online Event Registration Controller
def manage_event_registration(event_capacity, total_registrations):
    if total_registrations > event_capacity:
        confirmed = event_capacity
        waitlisted = total_registrations - event_capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
# Sample Run
manage_event_registration(100, 105)
```
