```python
# Problem 1: Social Media – Post Engagement Analyzer
# Create a function to calculate total likes and determine post status.

def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    # Check if post is viral
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example Data
likes = [200, 300, 150, 400]
analyze_post_engagement(likes)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
# Create a function to calculate average rainfall and check adequacy.

def check_rainfall(rainfall_data, required_level):
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    # Check rainfall adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example
rainfall = [60, 80, 70, 78]
check_rainfall(rainfall, 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
# Create a function to detect duplicate usernames using set.
def detect_duplicate_accounts(usernames):
    # Convert list to set
    unique_users = set(usernames)
    # Compare lengths
    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"
    
    print("Duplicate Accounts Found:", result)
# Example
users = ["rahul", "amit", "rahul", "neha"]
detect_duplicate_accounts(users)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
# Create a function to filter premium crop prices.
def filter_premium_crops(prices):
    premium_crops = []
    # Loop to filter premium crops
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)
# Example
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
# Create a function to evaluate system health based on error count.
def check_system_health(errors):
    
    # Conditional checks
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", errors)
    print("System Status:", status)
# Example
check_system_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
# Create a function to validate transaction amount against daily limit.
def check_transaction(amount):
    daily_limit = 50000  # Defined daily limit
    
    # Check transaction limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
# Example
check_transaction(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
# Create a function to calculate attendance percentage and exam eligibility.
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Count attended classes (1 means present)
    for status in attendance_list:
        if status == 1:
            attended_classes += 1
    percentage = (attended_classes / total_classes) * 100
    
    # Check eligibility
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

# Example
attendance = [1,1,1,0,1,1,1,1,0,1]
check_attendance(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
# Create a function to calculate electricity bill and classify usage.

def calculate_electricity_bill(units):
    bill = 0
    # Billing slab logic
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"    
    return bill, usage_status
# Example
total_bill, status = calculate_electricity_bill(250)
print("Total Bill:", total_bill)
print("Usage Status:", status)
```
