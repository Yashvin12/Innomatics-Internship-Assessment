```python
# Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Light is OFF by default
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def show_status(self):
        print(f"{self.name} is {self.status}")
# ---- Main Program ----
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)
if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()

light.show_status()
```

```python
# Problem 2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
# ---- Main Program ----
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
employee = Employee(name, emp_id, department)
employee.display_id_card()
```

```python
# Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def save_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

# ---- Main Program ----
contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")
contact = Contact(contact_name, phone_number)
contact.save_contact()
```

```python
# Problem 4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))
# ---- Main Program ----
product_name = input("Product Name: ")
price = input("Price: ")

product = Product(product_name, price)
product.print_price_tag()
```

```python
# Problem 5: Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def show_details(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")
# ---- Main Program ----
movie_name = input("Movie Name: ")
rating = input("Rating: ")
movie = Movie(movie_name, rating)
movie.show_details()
```

```python
# Problem 6: Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def show_delivery_details(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

# ---- Main Program ----
customer_name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(customer_name, address)
delivery.show_delivery_details()
```
