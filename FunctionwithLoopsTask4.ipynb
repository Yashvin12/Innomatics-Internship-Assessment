```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: ₹"))
        # Check minimum amount
        if amount < 50:
            print("Amount must be at least ₹50")
            continue
        # Check if plan is valid
        if amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Plan. Try again!")

# Function call
validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    # Loop through dictionary
    for product, stock in products.items():       
        # Check stock level
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")

# Sample data
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Salt": 18
}
check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def check_result(marks):
    total = 0
    
    # Calculate total marks
    for m in marks:
        total += m
    # Calculate average
    avg = total / len(marks)
    print("Average Marks:", avg)
    # Check pass/fail
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"
# Example
marks_list = [60, 45, 70, 55]
print(check_result(marks_list))
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    # Calculate normal fare
    fare = base_fare + (distance * per_km)
    # Add 25% extra during peak hours
    if peak == "yes":
        fare += fare * 0.25
    return fare

while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ")
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry != "yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance):
    present_days = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / len(attendance)) * 100
    print("Attendance %:", percentage)
    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance_list = ["P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance_list))
```
